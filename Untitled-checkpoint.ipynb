Read the data

```python
#按照 utf-8 编码的方式来读取程序，且以字符串形式读取。（后面会将 label 转为数字以便后续操作）
df_train = pd.read_csv("train.csv", encoding='utf-8').astype(str)
df_test = pd.read_csv("test.csv", encoding='utf-8').astype(str)
```

```python
#查看 train 集的前 5 条新闻。
df_train.head(5)
```

```python
#中文文本断词
def process(s):
    with open('stop_words.txt', 'r', encoding='utf-8') as f: # 打开停用词表
        stop_words = [line.strip() for line in f.readlines()]
    a = jieba.lcut(s)                                                         # 分词
    b = [w for w in a if w not in stop_words]                                 # 删除停用词及标点符号
    s = " ".join(b)
    return s
```

```python
#认为 News Url 和 Image Url 对判断新闻真假作用不大，删除这两列。
df_train=df_train.drop(["News Url","Image Url"],axis=1)
df_test=df_test.drop(["News Url","Image Url"],axis=1)
```

```python
#将 Ofiicial Account Name 和 Report Content 合并到 Title 列中，再删除 Ofiicial Account Name 和 Report Content 这两列。
#最后将合并好的特征调用分词函数，一同处理
df_train['Title'] = df_train['Title'] + df_train['Ofiicial Account Name']+df_train['Report Content']
df_train=df_train.drop(["Ofiicial Account Name","Report Content"],axis=1)
df_test['Title'] = df_test['Title'] + df_test['Ofiicial Account Name']+df_test['Report Content']
df_test=df_test.drop(["Ofiicial Account Name","Report Content"],axis=1)
```

```python
df_train["Title"] = df_train["Title"].apply(lambda x: process(x))
df_test["Title"] = df_test["Title"].apply(lambda x: process(x))
#df_train["Title"] = df_train["Title"].apply(lambda x: find_chinese(x))
#df_test["Title"] = df_test["Title"].apply(lambda x: find_chinese(x))
```

Building the RNN.

Convert text to vectors, our classifier only takes numerical data.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
x1=df_train["Title"]
x2=df_train["label"]
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(x1, x2, test_size=0.20, random_state=18)
```

```python
tokenizer=Tokenizer(num_words=10000)
##按单词出现次数排序，排序前10000的单词会列入词典中  
tokenizer.fit_on_texts(x1)
tokenizer.word_index
```

```python
#再将其转化为列表，这样就得到数字信息了。
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
```

```python
X_train
```

```python
X_test
```

```python
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
#再统一长度，保证每个文本都是同样的长度，避免不必要的错误。
X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, padding='post', maxlen=256)
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test, padding='post', maxlen=256)
```

```python
#搭建模型 Embedding 层将数字列表转化为向量列表
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 128),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64,  return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(16)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1)
])

model.summary()
```

```python
#EarlyStopping 当监测数量停止改善时停止训练。monitor：要监测的数据；val_loss
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)
model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(1e-4),
              metrics=['accuracy'])
#model.fit()方法来执行训练过程，每一批 batch 的大小为 30，迭代次数 epochs 为10，从测试集中划分 90%给训练集。
history = model.fit(X_train, Y_train, epochs=10,validation_split=0.1, batch_size=30, shuffle=True, callbacks=[early_stop])
```

```python
#fit 函数返回 History 对象，画出 history 中记录的各指标画出 loss 函数和 accuracy函数。
history_dict = history.history

acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']
epochs = history.epoch

plt.figure(figsize=(12,9))
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss', size=20)
plt.xlabel('Epochs', size=20)
plt.ylabel('Loss', size=20)
plt.legend(prop={'size': 20})
plt.show()

plt.figure(figsize=(12,9))
plt.plot(epochs, acc, 'g', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy', size=20)
plt.xlabel('Epochs', size=20)
plt.ylabel('Accuracy', size=20)
plt.legend(prop={'size': 20})
plt.ylim((0.5,1))
plt.show()
```

```python
model.evaluate(X_test, Y_test)
```

```python
pred = model.predict(X_test)

binary_predictions = []

for i in pred:
    if i >= 0.5:
        binary_predictions.append(1)
    else:
        binary_predictions.append(0) 
```

```python
print('Accuracy on testing set:', accuracy_score(binary_predictions, Y_test))
print('Precision on testing set:', precision_score(binary_predictions, Y_test))
print('Recall on testing set:', recall_score(binary_predictions, Y_test))
```

```python
#画出混淆矩阵 各块显示的是概率
import seaborn as sns 
matrix = confusion_matrix(binary_predictions, Y_test, normalize='all')
plt.figure(figsize=(8, 5))
ax= plt.subplot()
sns.heatmap(matrix, annot=True, ax = ax)

# labels, title and ticks
ax.set_xlabel('Predicted Labels', size=20)
ax.set_ylabel('True Labels', size=20)
ax.set_title('Confusion Matrix', size=20) 
ax.xaxis.set_ticklabels([0,1], size=15)
ax.yaxis.set_ticklabels([0,1], size=15)
```
